# Video Games Released in 2022

## About Dataset
The "Video Games Released in 2022" dataset is a comprehensive collection of information on video games that were released in the year 2022.

The Month column provides information on the month in which the game was released.

The Day column gives the day of the month when the game was released.

The Title column contains the title of the video game.

The Platform(s) column lists the platforms or devices on which the game is available, such as PlayStation, Xbox and Nintendo Switch.

The Genre(s) column lists the genre or type of the game, such as action, adventure, and sports.

The Developer(s) column lists the company or individual who developed the game.

The Publisher(s) column lists the company or individual who published the game.

This dataset is useful for anyone looking to research the video games released in 2022, to identify trends and patterns in the gaming industry, or to keep track of new releases for personal or professional use.

[Kaggle Dataset Link](https://www.kaggle.com/datasets/mattop/video-games-released-in-2022?resource=download)

## Data Visualization

Let's start to visualize the number of games that have been released each month.

```python
months = {"JANUARY":1,
          "FEBRUARY":2,
          "MARCH":3,
          "APRIL":4,
          "MAY":5,
          "JUNE":6,
          "JULY":7,
          "AUGUST":8,
          "SEPTEMBER":9,
          "OCTOBER":10,
          "NOVEMBER":11,
          "DECEMBER":12}
games["Month"] = games["Month"].map(months)
plt.figure(figsize=(10,6))
ax = plt.subplot(1, 1, 1)
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-dark.mplstyle')
plt.title("Number of Games Released in 2022 by Months")
ax = games["Month"].value_counts().sort_index().plot(kind="bar", rot=55, color=["magenta",
                                                                               "cornflowerblue",
                                                                               "cyan",
                                                                               "brown",
                                                                               "orangered",
                                                                               "gold",
                                                                               "limegreen",
                                                                               "mediumspringgreen",
                                                                               "lightseagreen",
                                                                               "dimgrey",
                                                                               "gray",
                                                                               "snow"])
ax.set_xticklabels(["JANUARY",
                    "FEBRUARY",
                    "MARCH",
                    "APRIL",
                    "MAY",
                    "JUNE",
                    "JULY",
                    "AUGUST",
                    "SEPTEMBER",
                    "OCTOBER",
                    "NOVEMBER",
                    "DECEMBER"])
ax.set_xlabel("Months")
ax.set_ylabel("Number of Games")
plt.show()
```

**Observation:** From the above graph, we can see that, in 2022, most games have been released during May, June, July, August and September.

Which developers are in the top 10 by number of games released?

```python
plt.figure(figsize=(10,6))
ax = plt.subplot(1, 1, 1)
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-dark.mplstyle')
plt.title("Number of Games Released in 2022 by Developers")
developers = games["Developer(s)"].apply(lambda x: [d.strip() for d in x.split("|")]).explode().value_counts()
developers = pd.Series(developers).where(lambda x: x >= 5).dropna()
developers.plot(kind="bar", rot= 90, color=["cornflowerblue", "cyan"])
ax.set_xlabel("Developers")
ax.set_ylabel("Number of Games")
plt.show()
```

**Obesrvation:** Square Enix and Nippon Ichi Software released 13 games followed by Inti Creates and Capcom with 8.

Which are the most popular genres between games that have been released in 2022?

```python
plt.figure(figsize=(10,6))
ax = plt.subplot(1, 1, 1)
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-dark.mplstyle')
plt.title("Most popular video games genres in 2022")
genres = games["Genre(s)"].apply(lambda x: [g.strip() for g in x.split("|")]).explode().value_counts()
genres = pd.Series(genres).where(lambda x: x >= 10).dropna()
genres.plot(kind="bar", rot= 90, color=["cornflowerblue", "cyan"])
ax.set_xlabel("Genres")
ax.set_ylabel("Number of Games")
plt.show()
```

**Observation:** The most popular genres among all the released games are: Adventure, Action role-playing, Role-playing, Visual Novel and Action-adventure.

- Now let's see which platforms have been the most popular in 2022

```python
platforms_dict = {}
for i in range(len(games)):
    platforms = [p.strip() for p in games.loc[i, "Platform(s)"].split(",")]
    for platform in platforms:
        if platform not in platforms_dict.keys():
            platforms_dict[platform] = 1
        else:
            platforms_dict[platform] += 1
df_platforms = pd.DataFrame(platforms_dict, index=["N. Platforms"]).T
popular_platforms = df_platforms[df_platforms["N. Platforms"] >= 50].sort_values(by="N. Platforms", ascending=False)

plt.figure(figsize=(10,6))
ax = plt.subplot(1, 1, 1)
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-dark.mplstyle')
plt.title("Most popular platforms in 2022")
sns.barplot(data=popular_platforms, x=popular_platforms.index, y=popular_platforms["N. Platforms"])
ax.set_xlabel("Platforms")
ax.set_ylabel("Number of Games")
plt.show()
```

**Observation:** The most popular platforms in 2022 are Nintendo Switch (NS) and PC(Win).
